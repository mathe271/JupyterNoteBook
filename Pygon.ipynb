```python
def circumscribed(r,n):
    a = 2 * r * sympy.tan(np.radians(180/n)) # 一辺 
    L = a * n # 外周
    Sp = 1/2 * n * a * r # 多角形面積
#     angle = 180 * (n-2)/n # 角度
    pi = L / 2 / r
    return a,L,Sp,pi
```

```python
def inscribed(r,n):
    a = 2 * r * sympy.sin(np.radians(180/n))
    L = a * n
    Sp = 1/2 * n * r **2 * sympy.sin(np.radians(360/n))
#     angle = 180 * (n-2)/n
    pi = L / 2 / r
    return a,L,Sp,pi
```

```python
def polyturtle(r,n,color):

    t.hideturtle()  
    turtle.hideturtle()
 
    turtle.pencolor(color)
    t.pencolor(color)
    
    t.pu()
    t.setpos(-175, 125)
    t.pd()
    t.write(arg=n, align="right",font=("Sans", 70, "bold"))
    
    turtle.pu()
    turtle.setpos(0, -1.5 * r)
    turtle.pd()
#     turtle.speed(0)
    
    a,L,Sp,pi = circumscribed(r,n)
            
    angle = 360 / n
    
#     turtle.circle(r)
    turtle.circle(r,steps=n)
    
    turtle.forward(a/2)
    
    for i in range(n-1):
        turtle.left(angle)
        turtle.forward(a)
        
    turtle.left(angle)
    turtle.forward(a/2)

    t.clear()
```

```python
fig, ax = plt.subplots(figsize=(6,6),subplot_kw={'aspect': 'equal'})
center = 2
ax.set_xticks([0, 1, 2, 3, 4])
ax.set_yticks([0, 1, 2, 3, 4])
ax.grid()
ax.set_axisbelow(True)

tbl = []

def update(n):
    
    cir = patches.Circle(xy=(center, center), radius=1,facecolor=None,edgecolor= "r",linewidth= 1,fill=False)

    tri = patches.CirclePolygon(xy=(center, center), radius=1, resolution=n, facecolor=None,edgecolor= "y",linewidth= 1,fill=False)

    r2 = r / (sympy.cos(np.radians(360/(2 * n))))

    outTri = patches.CirclePolygon(xy=(center, center), radius=r2, resolution=n, facecolor=None,edgecolor= "b",linewidth= 1,fill=False)

    art=ax.add_artist(tri)
    art=ax.add_artist(cir)
    art=ax.add_artist(outTri)
    
    ax.set_title("n="+str(n))
    
ani = FuncAnimation(fig, update, frames=range(3,30), repeat=False,interval = 500)
```
